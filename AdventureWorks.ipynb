### Importando as bibliotecas _pandas_ e _matplotlib.pyplot_ e definindo o estilo do _matplotlib.pyplot_ como sendo _seaborn_:

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

### Lendo o arquivo .xlsx e armazenando no Dataframe _df_:

```python
df = pd.read_excel("datasets\AdventureWorks.xlsx")
df
```

#### O Dataframe possui 904 linhas e 16 colunas

### Exibindo os tipos das colunas do Dataframe:

```python
df.dtypes
```

### Exibindo a soma dos valores da coluna "Valor Venda" com um arredondamento de 2 casas decimais:

```python
round(df["Valor Venda"].sum(), 2)
```

### Criando a coluna "Custo" através da multiplicação da coluna "Custo Unitário" pela coluna "Quantidade", como também fazendo a criação da coluna "Lucro" através da subtração da coluna "Valor Venda" pela coluna "Custo":

```python
df["Custo"]=df["Custo Unitário"].mul(df["Quantidade"])
df["Lucro"]=df["Valor Venda"] - df["Custo"]
df.head()
```

### Criando a coluna "Tempo de Entrega", formatada em número de dias com o tipo _int64_, através subtração da coluna "Data Envio" pela coluna "Data Venda":

```python
df["Tempo de Entrega"]=(df["Data Envio"]-df["Data Venda"]).dt.days
df
```

### Exibindo os tipos das colunas do Dataframe:

```python
df.dtypes
```

### Exibindo a média do Tempo de Entrega por Marca:

```python
df.groupby("Marca")["Tempo de Entrega"].mean()
```

### Alterando a opção do formato do _display_ do _float_ para 20 algarismos com 2 casas decimais:

```python
pd.options.display.float_format = "{:20,.2f}".format
```

### Criando um novo Dataframe "lucro_ano" através da soma do Lucro pelo agrupamento das colunas "Data Venda" e "Marca", e resetando o indíce da coluna:

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

### Exibindo a soma da Quantidade de Produtos Vendidos por Produto de forma decrescente:

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

### Plotando o gráfico de barras horizontais da soma da Quantidade de Produtos Vendidos por Produto de forma crescente:

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False).plot.barh(title="Total de Produtos Vendidos");
plt.xlabel("Total");
plt.ylabel("Produtos");
```

### Plotando o gráfico de barras verticais da soma do Lucro por Ano:

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro por Ano");
plt.xlabel("Ano");
plt.ylabel("Lucro");
```

### Plotando o gráfico de barras verticais da soma do Lucro por Ano com os indíces do eixo x escritos de forma horizontal:

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro por Ano");
plt.xlabel("Ano");
plt.ylabel("Lucro");
plt.xticks(rotation="horizontal");
```

### Exibindo informações estatísticas da coluna "Tempo de Entrega":

```python
df["Tempo de Entrega"].describe()
```

### Exibindo o gráfico _boxplot_ da coluna "Tempo de Entrega":

```python
plt.boxplot(df["Tempo de Entrega"]);
```
